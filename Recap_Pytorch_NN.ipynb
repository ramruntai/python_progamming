## Optimizers
### Optimizers are used to update weights and biases i.e. the internal parameters of a model to reduce the error

```python
import torch.optim as optim

params = torch.randn((3, 3), requires_grad=True)
learning_rate = 0.001

## SGD
optimizer = optim.SGD([params], lr=learning_rate)

# if we don't use optimizer, we have to manually update params
# for params in model.parameters(): 
#   params -= params.grad * learning_rate   

# We can use the step() method from our optimizer to take a forward step, instead of manually updating each parameter.
optimizer.step()
```

 ## Neural Network 
 #### Neural Network Construction: Create the neural network layers. setting up parameters (weights, biases)
 #### Forward Propagation: Calculate the predicted output. Measure error
 #### Back-propagation: After finding the error, then backward propagate our error gradient to update our weight parameters. This is doen by taking the derivative of the error function with respect to the parameters of our NN.
 #### Iterative Optimization: Minimize error as much as possible. \Kkeep updating the parameters iteratively by gradient descent.

### Step 1: Build Network Construction 

```python
# Defin the model. 

## inherit nn.Module. 
class SimpleModel(torch.nn.Module):
    def __init__(self):
        super(SimpleModel,self).__init__()
        self.hidden = torch.nn.Linear(1,10)
        self.output = torch.nn.Linear(10,1)
        self.relu = torch.nn.ReLU()
        
    def forward(self,x):
        hidden = self.hidden(x)
        activation = self.relu(hidden)
        output = self.output(activation)
        return output

model = SimpleModel()
#help(SimpleModel)
```

```python
print(model.hidden.weight)
print(model.hidden.bias)
```

### Step2. Forward propagation

```python
# Define the forward function in the model 
```

### Step3. Back-propagation 
#### Calcuate error or loss and backward propagate our error gradient to update our parameters. 

```python
input = torch.randn(1)
target= torch.randn(1)

output = model(input)
criterion = torch.nn.MSELoss()
loss = criterion(output, target)

print(model.hidden.bias.grad)

# Zeros the gradient buffers of all parameters
model.zero_grad() 

# propagate the error to update weight parameters. 
loss.backward()
print(model.hidden.bias.grad)
```

## Step 4: iterative Optimization 

```python
input = torch.randn(1)
target= torch.randn(1)

optimizer = torch.optim.Adam(model.parameters(), lr = 0.01)
optimizer.zero_grad() 
print(model.hidden.weight)
print(model.hidden.bias)

output = model(input) # calculate predict
loss = criterion(output, target) # calculate loss
loss.backward() # calculate gradient

optimizer.step() # update parameter 

print(model.hidden.weight)
print(model.hidden.bias)
```

## Step 2~3 

```python
import matplotlib.pyplot as plt # visualization
def training(n_epochs, optimizer,model,criterion, inputs, target):
    losses =[]
    for epoch in range(n_epochs):
        optimizer.zero_grad()
        outputs = model(inputs)
        loss = criterion(outputs,target)
        loss.backward()
        optimizer.step()
        losses.append(loss)
        
        #if epoch % 20 == 0:
        #   print('Epoch %d, Loss %f' % (epoch,loss))
    
    plt.plot(range(n_epochs),losses,'o')
    plt.ylabel('Loss')
    plt.xlabel('epoch')     
```

```python
x_t = torch.randn(300,1) # x_t = torch.randn(300)?
y_t = x_t.pow(2)

#print(x_t)
#print(y_t)

plt.plot(x_t,y_t,'o')
plt.xlabel('x')
plt.ylabel('y = x^2')
```

```python
print(model.hidden.weight)
print(model.hidden.bias)

training(n_epochs = 1500,optimizer = optimizer,model = model,criterion = criterion,inputs = x_t,target = y_t)

print(model.hidden.weight)
print(model.hidden.bias)
```

```python
x_test = torch.randn(300,1)
y = x_test.pow(2)+ 1.8*torch.rand(x_test.size())
y_hat = model(x_test).data.numpy()

#print(y)
#print(y_hat)

plt.clf()
plt.plot(x_test, y, 'go', label='True data', alpha=0.5)
plt.plot(x_test, y_hat, 'o', label='Predictions', alpha=0.5)
plt.legend(loc='best')
plt.show()
```
